```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
```

```python

df = pd.read_csv('/Users/denistkachenko/Documents/Facebook Ads Data.csv')
```

```python

numeric_columns = df.select_dtypes(include=['float64', 'int64'])
correlation_matrix = numeric_columns.corr()
```

```python
correlation_matrix
```

```python

plt.figure(figsize=(6, 5))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f", vmin=-1, vmax=1)
plt.title('  ')
plt.show()
```

```python

correlation_with_total_value = correlation_matrix['total_value'].drop('total_value').sort_values(ascending=False)
print("   total_value:")
print(correlation_with_total_value.head(1)) 
print("\n   total_value:")
print(correlation_with_total_value.tail(1))
```
