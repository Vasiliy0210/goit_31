```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt 
import seaborn as sns
import plotly.express as px
plt.rcParams['figure.figsize'] = [12, 8]
```

```python
df2 = pd.read_csv('~/Downloads/auto-mpg.csv')
df = sns.load_dataset('mpg')
```

```python
plt.figure(figsize=(12, 8))
plt.scatter(df2['horsepower'], 
            df2['mpg'], 
            c=df2['origin'], 
            cmap='viridis', 
            s=100,
            alpha=0.5, 
            edgecolors='black', 
            linewidths=1)
plt.tight_layout()
plt.xlabel('horsepower')
plt.ylabel('mpg')
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(20, 8))
plt.plot(df['horsepower'], df['mpg'], 'o', color='green', alpha=0.5, markeredgecolor='black', markeredgewidth=1)
plt.show()
```

```python
plt.figure(figsize=(20, 8))
plt.bar(df['origin'], df['mpg'], color={'green','blue','yellow' }, alpha=0.5, edgecolor='black', linewidth=1)
plt.xticks(rotation=45)
plt.xlabel('origin')
plt.show()
```

# Seaborn

```python
sns.set_style('darkgrid')
fig, ax = plt.subplots(figsize=(12, 8))
sns.lineplot(x='horsepower', 
             y='mpg', 
             hue='origin', 
             data=df, ax=ax, 
             palette='bright')
plt.xticks(rotation=45)
plt.show()
```

```python
sns.set_style('darkgrid')
fig, ax = plt.subplots(figsize=(22, 8))
data = sns.load_dataset('mpg')

graph = sns.lineplot(x='model_year', y='horsepower', hue="origin", data=data, palette='bright', ax=ax)
fig1 = graph.get_figure()
fig1.savefig('/Users/denistkachenko/Documents/lineplot.png')
```

```python
sns.set_style('darkgrid')
sns.stripplot(x='origin', y='mpg', hue='cylinders', data=df, size=10, edgecolor='black', linewidth=1)
```

```python
sns.set_style('darkgrid')
sns.set_context('poster')
fig, ax = plt.subplots(figsize=(12, 10))
sns.swarmplot(x='origin', y='mpg', hue='cylinders', data=df, size=8, edgecolor='black', linewidth=1, palette='bright')
sns.set_palette('Set3')
```

```python
sns.set_style('darkgrid')
fig, ax = plt.subplots(figsize=(12, 8))
sns.boxplot(x='origin', y='mpg', hue='cylinders', data=df, palette='bright')
plt.show()
```

```python
sns.set_style('darkgrid')
sns.violinplot(x='origin', y='mpg', hue='cylinders', data=df)
```

```python
fig = px.scatter(df, x='horsepower', y='mpg', color='origin', size='cylinders')
fig.show()
```

```python
fig = px.scatter(df, x='horsepower', y='mpg', color='origin', size='cylinders', marginal_y='violin', marginal_x='box')
fig.show()
```

```python
fig = px.scatter_3d(df, 
                    x='horsepower', 
                    y='mpg', 
                    z='model_year', 
                    color='origin', 
                    size='cylinders',
                    labels={'horsepower': 'HP', 'mpg': 'MPG', 'model_year': 'Year'},
                    )
fig['layout'].update(title='3D Scatter Plot', 
                     width=1100,
                     height=900,)
fig.show()
```
