```python
import statsmodels.stats.api as sms

# Parameters
baseline_cr = 0.17
mde_relative = 0.20  # 15% lift
target_cr = baseline_cr * (1 + mde_relative)

# Power analysis
effect_size = sms.proportion_effectsize(baseline_cr, target_cr)
required_n = sms.NormalIndPower().solve_power(
    effect_size,
    power=0.8,
    alpha=0.05,
    ratio=1
)

daily_users = 2000 * 0.34
days = (required_n * 2) / daily_users

print(f"Sample size per group: {round(required_n)}")
print(f"Total users needed: {round(required_n * 2)}")
print(f"Estimated days: {round(days, 1)}")
```
