```python
import numpy as np
import pandas as pd
import plotly.express as px

```

```python
applications = pd.read_csv('~/Downloads/applications(2.0).csv')
industries = pd.read_csv('~/Downloads/industries(2.0).csv')
```

```python
applications.info()
```

```python
applications.drop_duplicates(subset=['applicant_id'], inplace=True)
```

```python
applications['External Rating']=applications['External Rating'].fillna(0)
```

```python
applications['Education level'] = applications['Education level'].fillna('Середня')
```

```python
industries.info()
```

```python
data = pd.merge(applications, industries, on='Industry', how='left')
```

```python
data['Applied at'] = pd.to_datetime(data['Applied at'], format='mixed')
```

```python
data.info()
```

```python
weekdays = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday']
data['Applied day'] = data['Applied at'].apply(lambda x: x.day_name())
```

```python
data['Score'] = data['Score'] + data['Age'].between(35, 55).astype(int)*20 + (data['Applied day'].isin(weekdays))*20 + (data['Marital status'] == 'Married')*20 + (data['Location'].str.contains('Київ', na=False))*20 + (data['External Rating'] >=7)*20 - (data['External Rating'] <=2)*20 
```

```python
data['Score'] = np.select([data['External Rating']==0, data['Amount'].isna()], [0, 0], default=data['Score'])
```

```python
data
```

```python
data.drop(data[data['Score'] < 0].index  , inplace=True)
data['Score'] = np.where(data['Score'] > 100, 100, data['Score'])
```

```python
data['week'] = data['Applied at'].apply(lambda x: x.week)
data['week start'] = data['Applied at'].dt.to_period('W').apply(lambda r: r.start_time)
```

```python
import duckdb as db
```

```python
data = db.query("""
SELECT *,
    date_trunc('week', "Applied at") AS "week start"
    FROM data
""").to_df()
```

```python
data
```

```python
results = data.groupby('week start').agg('mean').reset_index()
```

```python
results
```

```python
fig = px.line(results, x='week start', y='Score')
fig.show()
```
