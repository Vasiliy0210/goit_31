```python
import pandas as pd
import numpy as np
import plotly.express as px
import plotly.graph_objects as go
from plotly.subplots import make_subplots
```

```python
df = pd.read_csv('~/Downloads/online_gaming_behavior_dataset.csv')
```

```python
df
```

```python
df2 = pd.read_csv("~/Downloads/raw_data_hw3.csv")
df2
```

```python
df2["payments"] = pd.to_numeric(
    df2["payments"].astype(str).str.replace(",", ".", regex=False),
    errors="coerce"
)
```

```python
median_per_column = df2.median(numeric_only=True)
```

```python
median_per_column
```

```python
fig = make_subplots(
    rows=2, cols=2,
    subplot_titles=('Age Distribution', 'In-Game Purchases Distribution',
                    'Play Time Hours Distribution', 'Avg Session Duration (Minutes) Distribution'),
    vertical_spacing=0.12
)

fig.add_trace(
    go.Box(y=df['Age'], name='Age', boxpoints='outliers'),
    row=1, col=1
)

fig.add_trace(
    go.Box(y=df2['payments'], name='In-Game Purchases', boxpoints='outliers'),
    row=1, col=2
)

fig.add_trace(
    go.Box(y=df['PlayTimeHours'], name='Play Time Hours', boxpoints='outliers'),
    row=2, col=1
)

fig.add_trace(
    go.Box(y=df['AvgSessionDurationMinutes'], name='Avg Session Duration', boxpoints='outliers'),
    row=2, col=2
)

fig.update_layout(
    height=800,
    title_text="Gaming Behavior Boxplots Analysis",
    showlegend=False
)

fig.update_yaxes(title_text="Age (years)", row=1, col=1)
fig.update_yaxes(title_text="Purchases (0/1)", row=1, col=2)
fig.update_yaxes(title_text="Hours", row=2, col=1)
fig.update_yaxes(title_text="Minutes", row=2, col=2)

fig.show()
```

```python
fig = px.box(df2, y="payments", points="outliers", title="In-Game Purchases (df2)")

fig.update_layout(
    height=800,
    width=1200,
    showlegend=False,
    margin=dict(l=60, r=40, t=60, b=60)
)

fig.update_yaxes(title_text="Payments")

fig.show()
```
