```python
school_teachers = ["Jane", "Jack", "James"]
```

```python
def school(students, *teachers):
    print(f"Hello Students: {students}")
    for teacher in teachers:
        print(f"Hello Teacher: {teacher}")
        
school("John", *school_teachers)
    
```

```python
def my_function(students, **teachers):
    print(f"Hello Students: {students}")
    for key, value in teachers.items():
        print(f"Hello Teacher: {key} {value}")

my_function(student1="John", teacher1="Jane", teacher2="Jack", teacher3="James")
```
