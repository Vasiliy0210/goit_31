```python
import pandas as pd
import duckdb
```

```python
df1 = pd.read_csv('/Users/denistkachenko/Documents/AccountsSF.csv')
df2 = pd.read_csv('/Users/denistkachenko/Documents/cohort_monthly.csv')
```

```python
main = duckdb.query("""

Select a."1-Year",
    b.sub_type,
    b.accountid

from df1 as a
left join df2 as b on b.accountid = a."1-Year"
where b.sub_type != 'Yearly'

""").df()

main
```

```python
main = duckdb.query("""

Select
    b.accountid,
    a."3-Month",
    b.sub_type
from df2 as b
left join df1 as a on b.accountid = a."3-Month"
where b.sub_type = '3-Month'
and b.cluster = 'Paid Pro Plan'
and a."3-Month" is null

""").df()

main
```
