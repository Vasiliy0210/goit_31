## 4. Data Cleaning

````python
# Create cleaned copies
metadata_clean = metadata.copy()
reviews_clean = reviews.copy()

print(f"Original metadata shape: {metadata.shape}")
print(f"Original reviews shape: {reviews.shape}")
````

````python
# Drop columns with too many missing values (>70%)
threshold = 0.7
metadata_clean = metadata_clean.loc[:, metadata_clean.isnull().mean() < threshold]
print(f"Columns after dropping high-missing columns: {metadata_clean.shape[1]}")
print(f"Dropped columns: {set(metadata.columns) - set(metadata_clean.columns)}")
````

````python
# Fill missing numerical values with median
metadata_clean['price_numeric'].fillna(metadata_clean['price_numeric'].median(), inplace=True)
metadata_clean['page_count'].fillna(metadata_clean['page_count'].median(), inplace=True)

# Fill categorical with mode or 'Unknown'
metadata_clean['language'].fillna('Unknown', inplace=True)
metadata_clean['format'].fillna('Unknown', inplace=True)

print("Missing values after cleaning:")
print(metadata_clean.isnull().sum()[metadata_clean.isnull().sum() > 0])
````

````python
# Remove duplicates
print(f"Duplicate rows in metadata: {metadata_clean.duplicated().sum()}")
print(f"Duplicate rows in reviews: {reviews_clean.duplicated().sum()}")

metadata_clean.drop_duplicates(inplace=True)
reviews_clean.drop_duplicates(inplace=True)

print(f"\nAfter removing duplicates:")
print(f"Metadata shape: {metadata_clean.shape}")
print(f"Reviews shape: {reviews_clean.shape}")
````

````python
# Handle outliers using IQR method for price
Q1 = metadata_clean['price_numeric'].quantile(0.25)
Q3 = metadata_clean['price_numeric'].quantile(0.75)
IQR = Q3 - Q1
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

outliers = metadata_clean[(metadata_clean['price_numeric'] < lower_bound) | 
                          (metadata_clean['price_numeric'] > upper_bound)]
print(f"Price outliers detected: {len(outliers)}")
print(f"Price range after outlier detection: ${lower_bound:.2f} to ${upper_bound:.2f}")

# Cap outliers instead of removing
metadata_clean['price_clean'] = metadata_clean['price_numeric'].clip(lower_bound, upper_bound)
````

## 5. Feature Engineering & Transformations

````python
# Create price categories
metadata_clean['price_category'] = pd.cut(metadata_clean['price_numeric'], 
                                           bins=[0, 10, 20, 30, float('inf')],
                                           labels=['Budget', 'Mid-range', 'Premium', 'Luxury'])

# Create page count categories
metadata_clean['length_category'] = pd.cut(metadata_clean['page_count'],
                                           bins=[0, 200, 400, 600, float('inf')],
                                           labels=['Short', 'Medium', 'Long', 'Very Long'])

print(metadata_clean[['title', 'price_numeric', 'price_category', 'page_count', 'length_category']].head(10))
````

````python
# Calculate price per page
metadata_clean['price_per_page'] = np.round(metadata_clean['price_numeric'] / metadata_clean['page_count'], 4)

print("Price per page statistics:")
print(metadata_clean['price_per_page'].describe())
````

````python
# Extract title length
metadata_clean['title_length'] = metadata_clean['title'].str.len()

# Check if subtitle exists
metadata_clean['has_subtitle'] = metadata_clean['subtitle'].notna().astype(int)

print(f"Books with subtitles: {metadata_clean['has_subtitle'].sum()}")
print(f"Average title length: {metadata_clean['title_length'].mean():.1f} characters")
````

## 6. Groupby & Aggregations

````python
metadata_clean
````

````python
# Group by language
language_stats = metadata_clean.groupby('language').agg({
    'title': 'count',
    'price_numeric': ['mean', 'median'],
    'page_count': 'mean'
}).round(2)

language_stats.columns = ['book_count', 'avg_price', 'median_price', 'avg_pages']
language_stats = language_stats.sort_values('book_count', ascending=False).head(10)
print("Top 10 Languages:")
print(language_stats)
````

````python
language_stats
````

````python
# Group by format
format_stats = metadata_clean.groupby('format').agg({
    'price_numeric': ['mean', 'min', 'max'],
    'page_count': 'mean',
    'title': 'count'
}).round(2)

format_stats.columns = ['avg_price', 'min_price', 'max_price', 'avg_pages', 'count']
format_stats = format_stats.sort_values('count', ascending=False)
print("Statistics by Format:")
print(format_stats.head(10))
````

````python
# Group by publisher - top publishers
publisher_stats = metadata_clean.groupby('publisher').agg({
    'title': 'count',
    'price_numeric': 'mean',
    'page_count': 'mean'
}).round(2)

publisher_stats.columns = ['books_published', 'avg_price', 'avg_pages']
publisher_stats = publisher_stats.sort_values('books_published', ascending=False).head(15)
print("Top 15 Publishers:")
print(publisher_stats)
````

````python
# Multiple groupby - price category and format
price_format_stats = metadata_clean.groupby(['price_category', 'format']).agg({
    'title': 'count',
    'page_count': 'mean'
}).round(0)

price_format_stats.columns = ['count', 'avg_pages']
price_format_stats = price_format_stats[price_format_stats['count'] > 50].sort_values('count', ascending=False)
print("Books by Price Category and Format (showing groups with >50 books):")
print(price_format_stats.head(20))
````

## 7. Joining/Merging DataFrames

````python
# Check common columns for joining
print("Metadata columns:", metadata_clean.columns.tolist()[:5])
print("Reviews columns:", reviews_clean.columns.tolist())

# Check if parent_asin is the join key
print(f"\nUnique parent_asin in metadata: {metadata_clean['parent_asin'].nunique()}")
print(f"Unique parent_asin in reviews: {reviews_clean['parent_asin'].nunique()}")
````

````python
# INNER JOIN - only books with reviews
books_with_reviews = pd.merge(metadata_clean, reviews_clean, 
                               on='parent_asin', 
                               how='inner',
                               suffixes=('_book', '_review'))

print(f"Inner join result: {books_with_reviews.shape}")
print(f"Columns: {books_with_reviews.columns.tolist()[:10]}...")
````

````python
# LEFT JOIN - all books, with reviews where available
all_books = pd.merge(metadata_clean, reviews_clean,
                     on='parent_asin',
                     how='left',
                     suffixes=('_book', '_review'))

print(f"Left join result: {all_books.shape}")
print(f"Books without reviews: {all_books['rating'].isna().sum()}")
````

````python
# OUTER JOIN - all books and all reviews
full_data = pd.merge(metadata_clean, reviews_clean,
                     on='parent_asin',
                     how='outer',
                     suffixes=('_book', '_review'))

print(f"Outer join result: {full_data.shape}")
print(f"Books without reviews: {full_data['title'].isna().sum()}")
print(f"Reviews without book metadata: {full_data['rating'].isna().sum()}")
````

````python
# Aggregate reviews per book
review_aggregates = reviews_clean.groupby('parent_asin').agg({
    'rating': ['mean', 'count', 'std'],
    'helpful_vote': 'sum',
    'verified_purchase': 'sum'
}).round(2)

review_aggregates.columns = ['avg_rating', 'review_count', 'rating_std', 'total_helpful_votes', 'verified_purchases']
review_aggregates = review_aggregates.reset_index()

print("Review aggregates shape:", review_aggregates.shape)
print(review_aggregates.head())
````

````python
# Join aggregated reviews with metadata
books_with_ratings = pd.merge(metadata_clean, review_aggregates,
                              on='parent_asin',
                              how='left')

# Fill NaN for books with no reviews
books_with_ratings['review_count'].fillna(0, inplace=True)
books_with_ratings['avg_rating'].fillna(0, inplace=True)

print(f"Books with ratings shape: {books_with_ratings.shape}")
print(books_with_ratings[['title', 'author_name', 'price_numeric', 'avg_rating', 'review_count']].head(10))
````

## 8. Advanced Analysis on Merged Data

````python
# Top rated books with significant reviews (>10)
top_rated = books_with_ratings[books_with_ratings['review_count'] > 10].nlargest(20, 'avg_rating')
print("Top 20 Rated Books (with >10 reviews):")
print(top_rated[['title', 'author_name', 'avg_rating', 'review_count', 'price_numeric']])
````

````python
# Most reviewed books
most_reviewed = books_with_ratings.nlargest(20, 'review_count')
print("Top 20 Most Reviewed Books:")
print(most_reviewed[['title', 'author_name', 'review_count', 'avg_rating', 'price_numeric']])
````

````python
# Correlation analysis
correlation_cols = ['price_numeric', 'page_count', 'avg_rating', 'review_count', 'price_per_page']
correlation_matrix = books_with_ratings[correlation_cols].corr()
print("Correlation Matrix:")
print(correlation_matrix)
````

````python
# Publisher performance analysis
publisher_performance = books_with_ratings.groupby('publisher').agg({
    'title': 'count',
    'avg_rating': 'mean',
    'review_count': 'sum',
    'price_numeric': 'mean'
}).round(2)

publisher_performance.columns = ['books', 'avg_rating', 'total_reviews', 'avg_price']
publisher_performance = publisher_performance[publisher_performance['books'] >= 10]
publisher_performance = publisher_performance.sort_values('avg_rating', ascending=False).head(20)

print("Top 20 Publishers by Average Rating (with >=10 books):")
print(publisher_performance)
````

## 9. Pivot Tables & Cross-tabulations

````python
# Pivot table: average price by format and price category
pivot_price = pd.pivot_table(books_with_ratings,
                              values='price_numeric',
                              index='format',
                              columns='price_category',
                              aggfunc='mean').round(2)

print("Average Price by Format and Price Category:")
print(pivot_price.head(10))
````

````python
# Pivot table: average rating by format and length category
pivot_rating = pd.pivot_table(books_with_ratings[books_with_ratings['review_count'] > 0],
                               values='avg_rating',
                               index='format',
                               columns='length_category',
                               aggfunc='mean').round(2)

print("Average Rating by Format and Book Length:")
print(pivot_rating.head(10))
````

````python
# Cross-tabulation: count of books by format and language
crosstab = pd.crosstab(books_with_ratings['format'], 
                       books_with_ratings['language'],
                       margins=True,
                       margins_name='Total')

print("Cross-tabulation: Books by Format and Language:")
print(crosstab.iloc[:10, :6])  # Show first 10 formats and 6 languages
````

## 10. Sorting & Ranking

````python
# Sort by multiple columns
sorted_books = books_with_ratings.sort_values(['avg_rating', 'review_count'], 
                                               ascending=[False, False])
print("Books sorted by rating and review count:")
print(sorted_books[['title', 'author_name', 'avg_rating', 'review_count', 'price_numeric']].head(15))
````

````python
# Add ranking columns
books_with_ratings['price_rank'] = books_with_ratings['price_numeric'].rank(ascending=True)
books_with_ratings['rating_rank'] = books_with_ratings['avg_rating'].rank(ascending=False, method='dense')
books_with_ratings['popularity_rank'] = books_with_ratings['review_count'].rank(ascending=False)

print("Books with rankings:")
print(books_with_ratings[['title', 'price_numeric', 'price_rank', 'avg_rating', 'rating_rank', 'review_count', 'popularity_rank']].head(10))
````

````python
# Percentile ranking
books_with_ratings['price_percentile'] = books_with_ratings['price_numeric'].rank(pct=True) * 100
books_with_ratings['rating_percentile'] = books_with_ratings['avg_rating'].rank(pct=True) * 100

print("Books with percentile rankings:")
print(books_with_ratings[['title', 'price_numeric', 'price_percentile', 'avg_rating', 'rating_percentile']].head(10))
````

## 11. Filtering & Querying

````python
# Multiple conditions filtering
high_value_books = books_with_ratings[
    (books_with_ratings['avg_rating'] >= 4.5) &
    (books_with_ratings['review_count'] >= 50) &
    (books_with_ratings['price_numeric'] <= 20)
]

print(f"High-value books (rating>=4.5, reviews>=50, price<=20): {len(high_value_books)}")
print(high_value_books[['title', 'author_name', 'avg_rating', 'review_count', 'price_numeric']].head(10))
````

````python
# Using query method
budget_paperbacks = books_with_ratings.query('format == "Paperback" and price_category == "Budget" and page_count > 200')
print(f"Budget paperback books (>200 pages): {len(budget_paperbacks)}")
print(budget_paperbacks[['title', 'price_numeric', 'page_count', 'avg_rating']].head(10))
````

````python
# isin filtering
popular_formats = ['Paperback', 'Hardcover', 'Kindle']
popular_format_books = books_with_ratings[books_with_ratings['format'].isin(popular_formats)]

print(f"Books in popular formats: {len(popular_format_books)}")
print(popular_format_books.groupby('format')['title'].count())
````

````python
# String filtering
fiction_books = books_with_ratings[books_with_ratings['title'].str.contains('Novel', case=False, na=False)]
print(f"Books with 'Novel' in title: {len(fiction_books)}")
print(fiction_books[['title', 'author_name', 'avg_rating', 'price_numeric']].head(10))
````

## 12. Window Functions & Rolling Calculations

````python
# Cumulative statistics by publisher
publisher_data = books_with_ratings.sort_values(['publisher', 'price_numeric'])
publisher_data['cumulative_books'] = publisher_data.groupby('publisher').cumcount() + 1
publisher_data['avg_price_so_far'] = publisher_data.groupby('publisher')['price_numeric'].cumsum() / publisher_data['cumulative_books']

sample_publisher = publisher_data[publisher_data['publisher'] == 'Penguin Books'].head(10)
print("Cumulative analysis for Penguin Books:")
print(sample_publisher[['title', 'price_numeric', 'cumulative_books', 'avg_price_so_far']])
````

````python
# Rolling average by sorted price
sorted_by_price = books_with_ratings.sort_values('price_numeric').reset_index(drop=True)
sorted_by_price['rolling_avg_price'] = sorted_by_price['price_numeric'].rolling(window=100, min_periods=1).mean()
sorted_by_price['rolling_avg_rating'] = sorted_by_price['avg_rating'].rolling(window=100, min_periods=1).mean()

print("Rolling averages (100-book window):")
print(sorted_by_price[['title', 'price_numeric', 'rolling_avg_price', 'avg_rating', 'rolling_avg_rating']].tail(20))
````

## 13. Advanced NumPy Operations

````python
# Create arrays for matrix operations
price_array = books_with_ratings['price_numeric'].values[:100]
rating_array = books_with_ratings['avg_rating'].values[:100]
page_array = books_with_ratings['page_count'].values[:100]

# Stack arrays
data_matrix = np.column_stack([price_array, rating_array, page_array])
print(f"Data matrix shape: {data_matrix.shape}")
print(f"First 5 rows:\n{data_matrix[:5]}")

# Matrix operations
means = np.mean(data_matrix, axis=0)
stds = np.std(data_matrix, axis=0)
print(f"\nColumn means: {means}")
print(f"Column standard deviations: {stds}")
````

````python
# Normalize data using NumPy (Z-score normalization)
normalized_data = (data_matrix - means) / stds
print("Normalized data statistics:")
print(f"Means after normalization: {np.mean(normalized_data, axis=0)}")
print(f"Stds after normalization: {np.std(normalized_data, axis=0)}")

# Min-max normalization
min_vals = np.min(data_matrix, axis=0)
max_vals = np.max(data_matrix, axis=0)
minmax_normalized = (data_matrix - min_vals) / (max_vals - min_vals)
print(f"\nMin-max normalized range: [{np.min(minmax_normalized):.2f}, {np.max(minmax_normalized):.2f}]")
````

````python
# Boolean indexing with NumPy
price_values = books_with_ratings['price_numeric'].values
rating_values = books_with_ratings['avg_rating'].values

# Find books that are both cheap and highly rated
cheap_mask = price_values < 15
highly_rated_mask = rating_values > 4.0
both_conditions = cheap_mask & highly_rated_mask

print(f"Cheap books: {np.sum(cheap_mask)}")
print(f"Highly rated books: {np.sum(highly_rated_mask)}")
print(f"Cheap AND highly rated: {np.sum(both_conditions)}")
print(f"Percentage: {np.sum(both_conditions) / len(price_values) * 100:.2f}%")
````
