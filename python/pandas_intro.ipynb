```python
data = customers.groupby('country')['account_id'].count()
data.drop('US', inplace=True)
data.reset_index()
data.index = coco.convert(names=data.index, to='name_short', not_found=None)
data
```

```python
fig = px.choropleth(data, locations=data.index,
                    color=data.values,
                    locationmode='country names',
                    color_continuous_scale=px.colors.sequential.Plasma
                )
fig['layout'].update(title='Customers by Country', width=1300, height=900,)
fig.show()
```
